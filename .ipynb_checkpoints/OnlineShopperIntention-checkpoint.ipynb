# Online Shoppers Intention

### Introduction:
Online shopping has become a significant segment in the retail industry, constituting a substantial portion of Business-to-Customer (B2C) revenue. Understanding the behaviors of online shoppers regarding their research and purchasing patterns holds immense importance for businesses. These insights enable targeted advertising, personalized marketing, and strategic deals to attract potential customers, thereby augmenting sales and overall revenue.

### Problem Statement:
Marketing teams continuously endeavor to optimize promotions, pricing strategies, personalization, and campaigns to enhance customer acquisition, retention, and revenue. However, pinpointing the most effective strategies poses a challenge. Leveraging machine learning algorithms to analyze past customer behavior and predict future outcomes based on diverse marketing strategies presents an opportunity.

The primary objective of this project is to develop a machine learning model capable of predicting whether a customer visiting an online shopping website will make a purchase. This prediction serves as a valuable tool for marketing teams to optimize promotions, pricing, and campaigns, thereby increasing the probability of purchase and subsequently augmenting revenue.

### Justification and Source of Dataset:
The dataset utilized for this project, named the "Online Shoppers Purchasing Intention Dataset," is obtained from the UCI Machine Learning Repository. It consists features from 12,330 unique sessions, belonging to a distinct user within a one-year period. This deliberate approach mitigates biases associated with particular marketing campaigns, special days, user profiles, or specific time periods.

Comprising various features that capture user behavior on an online shopping website, this dataset includes metrics such as the number of pages visited, visit duration, and traffic source type. Additionally, it encompasses a binary label that signifies whether the user completed a purchase.

This dataset is a valuable resource in addressing the earlier-mentioned problem statement, as it provides insights into the factors that influence user purchasing decisions in online shopping environments. Through comprehensive analysis, machine learning models can extract actionable insights and identify effective marketing strategies conducive to increasing the likelihood of purchase.

### Abstract and Business Objective:
The central objective revolves around identifying key metrics influencing a shopper's behavior and providing prioritized recommendations and performance enhancements. The primary attribute of interest is revenue, indicating whether a purchase was completed or not. The aim is to derive critical insights into user behavior patterns, allowing for informed decisions that drive sales optimization.

### Columns Descriptions:

- **Administrative:** Indicates the quantity of pages categorized as 'administrative' visited by the user.

- **Administrative_Duration:** Represents the time spent on pages falling into the 'administrative' category.

- **Informational:** Refers to the count of pages categorized as 'informational' that the user visited.

- **Informational_Duration:** Denotes the time spent on pages categorized as 'informational'.

- **ProductRelated:** Signifies the number of pages falling under the 'product related' type that the user accessed.

- **ProductRelated_Duration:** Reflects the time spent on pages categorized as 'product related'.

- **BounceRates:** Indicates the percentage of visitors who enter the website through a specific page and exit without engaging in any further activities.

- **ExitRates:** Represents the percentage of pageviews that conclude on a particular page within the website.

- **PageValues:** Signifies the average value of a page, considering the value of the target page and/or the completion of an eCommerce transaction.

- **SpecialDay:** A value that measures the proximity of a website visit to special events like Valentine's Day or Mother's Day, suggesting transaction likelihood. It accounts for the order-to-delivery duration, with values indicating the closeness to these occasions, possibly peaking on the day itself. The attribute is zero outside this window, except when near another special day.

- **Month:** Contains the month in which the pageview took place, presented in string format.

- **OperatingSystems:** An integer value that represents the operating system used by the user during their page viewing.

- **Browser:** An integer value that represents the browser employed by the user for viewing the pages.

- **Region:** An integer value that specifies the user's geographical region.

- **TrafficType:** An integer value categorizing the type of traffic the user is associated with.

- **VisitorType:** A string describing whether the visitor is a 'New Visitor,' 'Returning Visitor,' or falls into another category.

- **Weekend:** A boolean indicating whether the session occurred on a weekend.

- **Revenue:** A boolean value indicating whether or not the user completed a purchase.

## Data Cleaning

### Removing Features

Significant correlations were observed between certain pairs of features in the dataset. Specifically, strong correlations were found between "Administrative" and "Administrative_Duration," "Informational" and "Informational_Duration," "ProductRelated" and "ProductRelated_Duration," as well as between "ExitRates" and "BounceRates." To mitigate multicollinearity issues, a decision was made to remove one feature from each correlated pair. Furthermore, upon thorough data exploration, it became apparent that the "SpecialDay" feature lacked accuracy, leading to its removal from the dataset. 

```python
data_shopper = data.copy()
```

```python
data_shopper.drop(columns=['Administrative_Duration','Informational_Duration',
                           'ProductRelated_Duration','SpecialDay', 'BounceRates'], inplace=True)
```

### Handling outliers

A considerable presence of outliers was identified within the dataset. Upon thorough examination of density plots, extreme skewness was noted in two specific features, namely Informational_Duration and PageValues. Removing outliers from these features would result in an impractical scenario where only one value remains for each of these attributes.

To address this situation, a more selective approach to preprocessing was chosen. Instead of uniformly removing outliers across all features, a decision was made to selectively apply log transformation to certain features based on their individual characteristics. This approach enables outliers and skewness to be managed in a more focused manner.

Through this method, the aim is to preserve the dataset's integrity while effectively handling outliers. This strategy aligns with the project's approach, focusing on a feature-specific data preprocessing method to ensure the most suitable treatment for each variable.

```python
log1p_transform = FunctionTransformer(func=np.log1p, validate=False)
```

## Feature Engineering

After removing specific features, the dataset was organized into categorical and continuous columns.  This step is crucial as it sets the groundwork for preprocessing tasks, aiming to enhance the dataset's compatibility with various machine learning models intended for utilization later in the project.

```python
categorical_columns = ['Month', 'OperatingSystems', 'TrafficType', 'VisitorType','Browser','Region', 'Weekend']
continuous_columns = ['Administrative', 'Informational', 'ProductRelated',  'ExitRates', 'PageValues']
```

```python
data_shopper['Month'] = data_shopper['Month'].map({'Feb': 2, 'Mar': 3, 'May': 5,'June':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12})
data_shopper['VisitorType'] = data_shopper['VisitorType'].map({'Returning_Visitor':1, 'New_Visitor':2, 'Other':3})
data_shopper['Weekend'] = data_shopper['Weekend'].map({True:1, False:0})
```

```python
scaler = StandardScaler()
```

After the initial assessment, the plan to employ one-hot encoding for categorical variables was reconsidered due to the dataset containing numerous distinct categories. The implementation of one-hot encoding in such cases could potentially lead to an exponential increase in dimensions, impacting computational resources and model performance. To mitigate the issue of high dimensionality, label encoding was opted for categorical variables that were not inherently represented numerically. Label encoding assigns unique numerical labels to different categories within a feature, preserving valuable information without substantially increasing dimensionality.

Furthermore, for the continuous variables, standard scaling was employed. This scaling technique ensures that continuous features are standardized to a comparable scale, preventing variables with larger magnitudes from dominating the model. Standard scaling transforms the data to have a mean of zero and a standard deviation of one, enhancing the performance and stability of machine learning models when dealing with numerical data.

By effectively encoding categorical variables and scaling continuous ones, the dataset will be suitably preprocessed, ensuring its readiness for seamless integration into the chosen machine learning algorithms. This preparatory stage lays a solid foundation, enabling the models to yield more accurate predictions and analyses when applied to the refined dataset.

## Identify Training and Test Data

An 80-20 split was employed to split the dataset into training and testing sets. This approach reserved 80% of the data for training the model, while the remaining 20% was designated for evaluating its performance. This division allowed the model to learn patterns from the larger training set while maintaining a separate segment for assessing its generalization to new, unseen data.

```python
train_set,test_set = train_test_split(data_shopper, test_size=0.2, random_state=42)


X_train = train_set.drop('Revenue', axis=1)
y_train = train_set['Revenue'].map({True:1,False:0})

X_test = test_set.drop('Revenue', axis=1)
y_test = test_set['Revenue'].map({True:1,False:0})
```

### Pipeline for data preprocessing

A pipeline was developed to streamline the preprocessing of continuous variables, ensuring consistency and reproducibility in data preparation. This pipeline was then separately applied to both the training and testing datasets. The rationale behind this separation lies in the necessity to prevent data leakage and maintain the integrity of model evaluation. By preprocessing the training and testing data independently, it ensures that the testing data remains untouched during the preprocessing stage, mirroring real-world scenarios where the model is deployed on new, unseen data. This separation minimizes the risk of the model learning from the testing data, which could compromise its generalizability and result in overly optimistic performance metrics. 

```python
num_pipeline = Pipeline([
    ('log_transform', log1p_transform),
    ('std_scaler', scaler)
])

X_train_preprocessed = num_pipeline.fit_transform(X_train)
X_test_preprocessed = num_pipeline.fit_transform(X_test)
```

# ML Algorithms

## Training the Data

### Naive Bayes

```python
nb_model = BernoulliNB()
nb_scores = cross_val_score(nb_model, X_train_preprocessed, y_train, cv=5)
print(nb_scores)
```

The Naive Bayes Classifier (BernoulliNB) showed good performance initially, with cross-validation scores ranging from around 85.9% to 87.2%. This classifier worked well without any adjustments, demonstrating its reliability in making predictions.

#### Hyperparameter Tuning - Naive Bayes

```python
param_grid = {
    'alpha': [0.1, 0.5, 1.0],
    'binarize': [0.0, 0.1, 0.5, 1.0] 
}

grid_search_bernoulli_nb = GridSearchCV(nb_model, param_grid, scoring='accuracy', cv=5, n_jobs=-1, verbose=1)
grid_search_bernoulli_nb.fit(X_train_preprocessed, y_train)

print("Best hyperparameters found:")
for key, value in grid_search_bernoulli_nb.best_params_.items():
    print(f"{key}: {value}")

print(f"Best accuracy: {grid_search_bernoulli_nb.best_score_}")
```

After applying hyperparameter tuning using GridSearchCV, the Naive Bayes Classifier (BernoulliNB) improved further. With this process, the best parameters—alpha set at 0.1 and a binarization threshold of 1.0—were identified. Following the optimization, the accuracy increased notably to around 88.3%. This shows that refining the parameters enhanced the classifier's accuracy, emphasizing the importance of adjusting settings to improve model performance.

###  SVM Classification

```python
svm_model = SVC()
svm_scores = cross_val_score(svm_model, X_train, y_train, cv=5, scoring='accuracy')
print(svm_scores)
```

Initially configured with default settings, the SVM classifier exhibited consistent and robust performance across various validation folds. Its accuracy scores, consistently hovering between 88.7% to 90.3%, reflected the model's inherent ability to make accurate predictions without any adjustments made to its default parameter values. This inherent strength in accuracy highlighted the SVM's effectiveness in handling the dataset with its default configurations.

#### Hyperparameter Tuning - SVM Classification

```python
param_grid = {
    'kernel': ['linear', 'rbf'],
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto', 0.1, 1, 10]
}

grid_search_svm = GridSearchCV(svm_model, param_grid, scoring='accuracy', cv=5, n_jobs=-1, verbose=1)
grid_search_svm.fit(X_train_preprocessed, y_train)

print("Best hyperparameters found:")
for key, value in grid_search_svm.best_params_.items():
    print(f"{key}: {value}")

print(f"Best accuracy: {grid_search_svm.best_score_}")
```

Following an extensive GridSearchCV for hyperparameter optimization with C=1, gamma='scale', and kernel='rbf', the SVM classifier's accuracy slightly increased to approximately 90.0%. This improvement, although higher than the lower end of the initial accuracy range (88.7%), still falls within the spectrum observed without parameter adjustments. Even with these adjusted settings, there's only a small improvement in accuracy, suggesting that tweaking the parameters didn't drastically enhance the model's performance.

### Decision Tree

```python
dt_model = DecisionTreeClassifier()
dt_scores = cross_val_score(dt_model, X_train_preprocessed, y_train, cv=5, scoring='accuracy')
print(dt_scores)
```

The initial Decision Tree model, utilizing default settings, displayed accuracy scores ranging from around 85.7% to 87.1% across five cross-validation folds. #### Hyperparameter Tuning - Decision Tree

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [25, 30, 35, 40],
    'max_features': ['sqrt', 'log2'],
    'max_leaf_nodes': [10, 30, 50],
    'min_impurity_decrease': [0.01, 0.1, 0.3, 0.5],
    'class_weight': ['balanced']
}

grid_search_dt = GridSearchCV(dt_model, param_grid, scoring='accuracy', cv=5, n_jobs=-1, verbose=1)
grid_search_dt.fit(X_train_preprocessed, y_train)

print("Best hyperparameters found:")
for key, value in grid_search_dt.best_params_.items():
    print(f"{key}: {value}")

print(f"Best accuracy: {grid_search_dt.best_score_}")
```

Conducting GridSearchCV for hyperparameter tuning resulted in an optimized Decision Tree model with the following parameters: criterion='gini', max_depth=40, max_features='log2', max_leaf_nodes=10, min_impurity_decrease=0.01, and class_weight='balanced'. This tuning process led to a minor accuracy improvement to about 87.4%, showcasing a slight enhancement in performance due to fine-tuning the parameters. Despite exploring an extensive parameter space with 192 candidates, the model's accuracy improvement remained marginal, indicating potential limitations in boosting accuracy through parameter adjustments for this dataset.
